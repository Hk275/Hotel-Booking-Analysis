```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("hotel_bookings.csv")
```

```python
# EDA and Data Cleaning

df.head()

```

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
# converting to date time 
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'], dayfirst=True)``
```

```python
df.describe(include='object')
```

```python
for col in df.describe(include='object').columns:
    print(col)
    print(df[col].unique())
    print('-'*50)
```

```python
df.isnull().sum()
```

```python
# Impossible to handle agent (not important) and company columns (too many), so we will drop them 
df.drop(['agent', 'company'], axis=1, inplace=True)
df.dropna(inplace=True)
```

```python
df.isnull().sum()   
```

```python
df.describe()
```

```python
# Checking outlier for adr 
df['adr'].plot(kind='box')
```

```python
# remove outlier for average daily rate
df= df[df['adr']<5000]
```

```python
# Now lets analyse the clean data
cancelled_perc = df['is_canceled'].value_counts(normalize=True)*100
print(cancelled_perc)
plt.figure(figsize=(10,6))
plt.title('Reservation Status Count') 
plt.bar(['Not Cancelled', 'Cancelled'], df['is_canceled'].value_counts(), edgecolor = 'k', width=0.7)
plt.show()
```

```python
plt.figure(figsize=(10,6))
ax1 = sns.countplot(x='hotel', hue = 'is_canceled', data = df, palette='viridis')
ax1.legend(bbox_to_anchor=(1, 1))
plt.title('Reservation Status in Different Hotels', size =20 )
plt.xlabel('Hotel')
plt.ylabel('Number of Reservations')
plt.legend(['Not Cancelled', 'Cancelled'])
plt.show()
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)*100
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)*100
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
city_hotel['adr']
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in Resort Hotel and City Hotel', size =20)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
ax1 = sns.countplot(x='month', hue='is_canceled', data=df, palette='viridis')
plt.title('Reservation Status in Different Months', size=20)
plt.xlabel('Month')
plt.ylabel('Number of Reservations')
plt.legend(['Not Cancelled', 'Cancelled'])
plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.title('ADR per month', size=30)
sns.barplot(x='month', y='adr', data=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index() , palette='viridis')
plt.show()
```

```python
cancelled_data = df[df['is_canceled'] == 1]
top_10_countries = cancelled_data['country'].value_counts().head(10)
plt.figure(figsize=(8,8))
plt.pie(top_10_countries, labels=top_10_countries.index, autopct='%1.1f%%', startangle=90)
plt.title('Top 10 Countries with Most Cancellations', size=20)
plt.show()
```

```python
df['market_segment'].value_counts()
```

```python
df['market_segment'].value_counts(normalize=True)*100
```

```python
cancelled_data['market_segment'].value_counts(normalize=True)*100
```

```python
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

not_cancelled_data = df[df['is_canceled'] == 0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in Cancelled and Not Cancelled Reservations', size =20)
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label = 'Not Cancelled')

plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label = 'Cancelled')

plt.legend(fontsize=20)
```

```python
cancelled_df_adr_1 = cancelled_df_adr[(cancelled_df_adr['reservation_status_date'] > '2016')]
# cancelled_df_adr_1.head()
not_cancelled_data = not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date'] > '2016') ]
not_cancelled_data.head()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in Cancelled and Not Cancelled Reservations', size =20)
plt.plot(not_cancelled_data['reservation_status_date'], not_cancelled_data['adr'], label = 'Not Cancelled')
plt.plot(cancelled_df_adr_1['reservation_status_date'], cancelled_df_adr_1['adr'], label = 'Cancelled')
plt.legend(fontsize=20)
plt.show()
```
